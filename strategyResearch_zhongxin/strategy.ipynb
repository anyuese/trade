```python
def caculate_turnover_rate(stock_pre,stock_now):
    stock_common = list(set(stock_pre).intersection(stock_now))
    print('公共股票：',stock_common)
    turnover_rate = (len(stock_pre)-len(stock_common))/len(stock_pre) * 0.004
    print('换手率：',turnover_rate)
    return turnover_rate
```
